# FairPrice Auto: EDA and Modeling

This notebook explores the cleaned used car dataset, reviews model comparison results, and examines the fairness label distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
processed_path = '../data/processed/used_cars_processed.csv'
df = pd.read_csv(processed_path)
df.head()
```

```python
df.info()
df.describe(include='all').transpose()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=25, kde=True, color='teal')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
if 'odometer' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='odometer', y='price', data=df, alpha=0.7)
    plt.title('Price vs Mileage')
    plt.xlabel('Odometer')
    plt.ylabel('Price')
    plt.show()
```

```python
metrics_path = '../results/model_metrics.csv'
metrics = pd.read_csv(metrics_path)
metrics
```

```python
predictions_path = '../results/predictions.csv'
predictions = pd.read_csv(predictions_path)
predictions.head()
```

```python
if 'fairness_label' in predictions.columns:
    predictions['fairness_label'].value_counts().plot(kind='bar', figsize=(8, 5), color=['gold', 'lightgreen', 'salmon'])
    plt.title('Fairness Label Distribution')
    plt.xlabel('Fairness Label')
    plt.ylabel('Count')
    plt.show()
```
